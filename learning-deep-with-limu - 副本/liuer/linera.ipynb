```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator

x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]

def forward(x,w,b):
    return x*w+b

def loss(y_h,y):
    return (y_h-y)**2

w_list = []
b_list = []
mse_list = []

for w_i in np.arange(0.0,4.1,0.1):
    for b_i in np.arange(-2.0,2.0,0.1):
        l_sum = 0.0
        for x_val,y_val in zip(x_data,y_data):
            y_h = forward(x_val,w_i,b_i)
            l = loss(y_h,y_val)
            l_sum += l
        mse_list.append(l_sum/len(y_data))
        w_list.append(w_i)
        b_list.append(b_i)

w_array = np.array(w_list)
b_array = np.array(b_list)
mse_array = np.array(mse_list)

W,B = np.meshgrid(np.arange(0.0,4.1,0.1),np.arange(-2.0,2.0,0.1))
MSE = mse_array.reshape(W.shape)

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
surface = ax.plot_surface(W,B,MSE,cmap='viridis',edgecolor='none')

ax.set_xlabel("W")
ax.set_ylabel("Y")
ax.set_zlabel("MSE")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

x_data = np.array([1.0, 2.0, 3.0])
y_data = np.array([2.0, 4.0, 6.0])

def forward(x, w, b):
    return x * w + b

def loss(y_h, y):
    return (y_h - y) ** 2

w_list = []
l_list = []

for w in np.arange(0.0,4.1,0.1):
    l_sum = 0.0
    for x_val,y_val in zip(x_data,y_data):
        y_h = forward(x_val,w,0.0)
        l_val = loss(y_h,y_val)
        l_sum += l_val
    l_list.append(l_sum/len(y_data))     
    w_list.append(w)

plt.plot(w_list,l_list)
plt.ylabel("mse")
plt.xlabel("loss")
plt.show()
```
