```python
import numpy as np
import matplotlib.pyplot as plt

x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = 1.0



def forward(x):
    return w*x


def loss(x, y):
    cost = 0
    for x_i, y_i in zip(x, y):
        y_h = forward(x_i)
        l = (y_h-y_i)**2
        cost += l
    return cost/len(x)


def gradient(x_list, y_list):
    g_sum = 0
    for x_i, y_i in zip(x_list, y_list):
        g = 2*x_i*(w*x_i-y_i)
        g_sum += g
    return g_sum/len(x_list) 

lr = 0.01
epoch_list = []
loss_list = []
for epoch in range(200):
    cost_val = loss(x_data, y_data)
    gradient_val = gradient(x_data, y_data)
    w = w - lr * gradient_val
    epoch_list.append(epoch)
    loss_list.append(cost_val)
    print(f"{epoch}:cost is {cost_val} w is {w}")

print(f"if x= 4 then y get {forward(4)}")

plt.plot(epoch_list,loss_list)
plt.xlabel("epoch")
plt.ylabel("cost")
plt.show()

```

接下来我们使用随机梯度算法，不再对一个整个数据集进行求梯度，而是对数据集中的每一个数据都求一遍梯度

```python
import numpy as np
import matplotlib.pyplot as plt

x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

w = 1.0



def forward(x):
    return w*x


def loss(x, y):
    y_h = forward(x)
    l = (y_h-y)**2
    return l


def gradient(x, y):
    g = 2*x*(w*x-y)
    return g 

lr = 0.01
epoch_list = []
loss_list = []
for epoch in range(200):
    for x,y in zip(x_data,y_data):
        l = loss(x,y)
        gradient_val = gradient(x,y)
        w = w - lr * gradient_val
        loss_list.append(l)
        epoch_list.append(epoch)
    print(f"{epoch}:cost is {cost_val} w is {w}")

print(f"if x= 4 then y get {forward(4)}")

plt.plot(epoch_list,loss_list)
plt.xlabel("epoch")
plt.ylabel("cost")
plt.show()

```
